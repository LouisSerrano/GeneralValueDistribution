```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
def gaussianKernel(x, y, sigmaK = 0.5):
    d = np.sum(x*y + y*y - 2*x*y)
    return np.exp(-d/(2*sigmaK**2))
```

```python
def muGaussian(x, n = 20000, mean = 0, sigma = 1, kernel = "gaussian"):
    dist = st.norm(mean, sigma)
    sample = dist.rvs(size = n)
    matrix = np.array([[gaussianKernel(s, y) for s in sample] for y in x])
    
    return np.mean(matrix, axis = 1)
```

```python
def muLaplace(x, n = 20000, mean = 0, sigma = 1, kernel = "gaussian"):
    dist = st.laplace(mean, sigma)
    support = dist.rvs(size = n)
    matrix = np.array([[gaussianKernel(s, y) for s in sample] for y in x])
    return np.mean(matrix, axis = 1)
```

```python
s = np.arange(-6,6,0.05)
```

```python
plt.plot(s, muGaussian(s))
```
